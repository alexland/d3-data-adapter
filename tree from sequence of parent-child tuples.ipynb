```python
from functools import wraps

def memoize(fn):
    cache = {}
    @wraps(fn)
    def wraps(*args):
        if not args in cache:
            cache[args] = fn(*args)
        return cache[args]
    return wraps
```

### Building the JSON Tree

```python
LuT = {}

def hash_fn(x):
    if x not in LuT.keys():
        LuT[x] = str(hash(x))[:10]
    return LuT[x]


# data = [list(map(hash_fn, inner_seq)) for inner_seq in data]
# data = list(id_pairs)
```

```python
'''
schema: tree data is in the form of a sequence of (parent, child) 2-tuples;
    no root node is specified in the data
'''

import json as JSON
import os

tmp = {}
my_tree = dict(name='r', children=[])
n = 0

for p, c in data:
    n += 1
    pn = tmp.get(p)   # has this parent node been seen before? 
    if not pn:
        # node doesn't exist in tree so add it to tmp
        # & update tree by adding this node as a 'parent'
        # to 'children' of root
        tmp[p] = pn = {'name': p}
        my_tree['children'].append(pn)
    tmp[c] = cn = {'name': c}
    # if 'children' in pn, return its value, otherwise, 
    # insert 'children' key w/ value of []
    # in either case, what is returned from 'setdefault' is a list
    # & to that list is appended cn
    pn.setdefault('children', []).append(cn)

print(len(data))
```

### Persist Tree to a JSON file

```python
print(JSON.dumps(my_tree, indent=4))
viz_file = os.path.expanduser("~/Dropbox/dict-tree-current/tx-0.json")
with open(viz_file, 'w', encoding='utf-8') as fh:
    JSON.dump(my_tree, fh)
```
